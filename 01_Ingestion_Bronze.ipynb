## Creating 'Bronze' database for data ingestion purposes.

```python
%sql
CREATE DATABASE IF NOT EXISTS bronze;
```

## Ingestion of raw data into 'Bronze' layer

```python
%sql

DROP TABLE IF EXISTS bronze.website_sessions;
DROP TABLE IF EXISTS bronze.website_pageviews;
DROP TABLE IF EXISTS bronze.order_item_refunds;
DROP TABLE IF EXISTS bronze.order_items;
DROP TABLE IF EXISTS bronze.orders;
DROP TABLE IF EXISTS bronze.products;

CREATE TABLE bronze.website_sessions
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'website_sessions.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/website_sessions.csv`
WITH (header="true", inferSchema="true");

CREATE TABLE bronze.website_pageviews
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'website_pageviews.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/website_pageviews.csv`
WITH (header="true", inferSchema="true");

CREATE TABLE bronze.order_item_refunds
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'order_item_refunds.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/order_item_refunds.csv`
WITH (header="true", inferSchema="true");

CREATE TABLE bronze.order_items
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'order_items.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/order_items.csv`
WITH (header="true", inferSchema="true");

CREATE TABLE bronze.orders
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'orders.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/orders.csv`
WITH (header="true", inferSchema="true");

CREATE TABLE bronze.products
USING DELTA
AS
SELECT 
    *,
    current_timestamp() AS _ingested_at,
    'products.csv' AS _source_file
FROM csv.`/Volumes/data_sets/default/maven-fuzzy-factory-data/products.csv`
WITH (header="true", inferSchema="true");
```
